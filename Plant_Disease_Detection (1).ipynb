```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")

```

```python
model.train(
    data="/content/Fruits-Diseases-Detection-1/data.yaml",
    epochs=15,
    imgsz=640,
    batch=16,
    name="fruits-diseases-detection"
)
```

```python
# Load the best model
trained_model = YOLO('/content/runs/detect/fruits-diseases-detection2/weights/best.pt')  # Update with your path
```

```python
test_image="/content/runs/detect/fruits-diseases-detection2/val_batch1_pred.jpg"
results = trained_model(test_image)
for result in results:
    result.plot()
```

```python
metrics = model.val()  # Runs validation
```
